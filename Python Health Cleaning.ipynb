```python
## Loading Librarbies and Checking For Availability
try:
    import pandas as pd
    print("Library is available.")
    from sklearn.preprocessing import LabelEncoder
    print("Library is available.")
    
except ImportError:
    print("Library is not available.")

# Changing Potential Options
#pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
## Loading Dataset File To Clean
health = pd.read_excel('Health Data.xlsx')
 
health.head(10)
```

```python
## Analyzing Structure of Data
health.info()
```

```python
## Changing Names For Simplier Use
health.rename(columns = {'Date': 'Date', 'Weight (Pounds)': 'Weight', 'Waist (Inches)': 'Waist', 'Neck (Inches)': 'Neck', 
                         'Morning Body Temp': 'Morning Body Temp', 'Morning Systolic Pressure': 'Morning Systolic Pressure', 
                         'Morning Diastolic Pressure': 'Morning Diastolic Pressure', 'Morning Pulse': 'Morning Pulse', 'Night Body Temp': 'Night Body Temp', 
                         'Night Systolic Pressure': 'Night Systolic Pressure', 'Night Diastolic Pressure': 'Night Diastolic Pressure', 
                         'Night Pulse': 'Night Pulse', 'Sleep': 'Sleep', 'BMI': 'BMI', 'CBF': 'CBF', 'Gym': 'Gym', 'Cardio': 'Cardio', 
                         'Calories': 'Calories', 'Carbs': 'Carbs', 'Fat': 'Fat', 'Protein': 'Protein', 'Fiber': 'Fiber', 'Sugar': 'Sugar', 
                         'Servings': 'Servings', 'Daily Water Intake (liters)': 'Water', 'Fat Calories': 'Fat Calories'}, inplace=True)
```

```python
## Confirming Name Changes
for i in health:
    print(i)
```

```python
## Changing Variables To Binary
le = LabelEncoder()
health['Gym'] = le.fit_transform(health['Gym'])
health['Cardio'] = le.fit_transform(health['Cardio'])
print(health[['Gym', 'Cardio']])
```

```python
## Changing Data Types and Rounding to 2 Decimal Places
columns = health[health.columns[1:26]]
for i in columns:
    health[i] = health[i].astype(float)
    health[i] = health[i].round(2)
    print(health[i].head(10))
```

```python
## Confirming Type Change
health.info()
```

```python
## Check for Missing Values or Null Values
print(health.isnull().sum())

## Counting Total Missing Values
print("\nTotal Missing:", health.isnull().sum().sum())
```

```python
## Confirming Missing Values
health[health.columns[17:26]].iloc[211:218]
```

```python
## Documenting Mean Values
meanbefore = health[health.columns[17:26]].mean()
print(meanbefore)
```

```python
## Replacing Null Values With Mean
columns = health.columns[17:26]
for column in columns:
    health[column] = health[column].fillna(health[column].mean())
```

```python
## Confirming Mean Replacement
health[health.columns[17:26]].iloc[211:218]
```

```python
## Comparing Current to Previous Means
meanafter = health[health.columns[17:26]].mean()
print("Mean Before:", "\n", meanbefore)
print("Mean After:", "\n", meanafter)
```

```python
## Save the DataFrame to an Excel file
health.to_excel('health_cleaned_python.xlsx', index=False)
```
